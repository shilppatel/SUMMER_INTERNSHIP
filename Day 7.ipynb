```python
from __future__ import division
from math import log


def entropy(pi):
    '''
    return the Entropy of a probability distribution:
    entropy(p) = − SUM (Pi * log(Pi) )
    defintion:
            entropy is a metric to measure the uncertainty of a probability distribution.
    entropy ranges between 0 to 1
    Low entropy means the distribution varies (peaks and valleys).
    High entropy means the distribution is uniform.
    See:

    '''

    total = 0
    for p in pi:
        p = p / sum(pi)
        if p != 0:
            total += p * log(p, 2)
        else:
            total += 0
    total *= -1
    return total


def gain(d, a):
    '''
    return the information gain:
    gain(D, A) = entropy(D)−􏰋 SUM ( |Di| / |D| * entropy(Di) )
    '''

    total = 0
    for v in a:
        total += sum(v) / sum(d) * entropy(v)

    gain = entropy(d) - total
    return gain



# set of example of the dataset
survived = [342, 549] # Yes, No


# attribute, number of members (feature)
pclass = [
    [136, 80],  # 1
    [87, 97],  # 2
    [119, 372],  # 3
    
]
sex = [
    [109, 468],  # male
    [233, 81],  # female
    
]
embarked = [
    [217, 427],  # 0 S
    [93, 75],   # 1 C
    [32, 47]   # 2 Q
]
parch = [
    [233, 445],  # 0
    [65, 53],   # 1
    [40, 40],  # 2
    [3, 2],  # 3
    [0, 4],  # 4
    [1, 4],   # 5
    [0, 1],   # 6
]

SibSp = [
    [210, 398],  # 0
    [112, 97],   # 1
    [13, 15],  # 2
    [4, 12],  # 3
    [3, 15],  # 4
    [0, 5],   # 5
    #[0, 0],   # 6
   # [0, 0],   # 7
    [0, 7],   # 8
]

print("pclass",gain(survived, pclass))
print("sex",gain(survived, sex))
print("embarked",gain(survived, embarked))
print("parch",gain(survived, parch))
print("SibSp",gain(survived, SibSp))
```
